```python
import pandas as pd
import numpy as np
from p0_common import run_label_percentile
```

```python
df = pd.read_csv('__filtered_result__\\normalies_df_last_two_eth_pc_change.csv')       #enable this for last two txn
#df = pd.read_csv('__filtered_result__\\normalies_df_f2l_eth_pc_change.csv')            #enable this for first to last txn
df = df.sort_values(['eth_pc_change'], ascending=True)
```

```python
percentile_bins = [0,25,50,75,100]
percentile_cutoffs = np.percentile(df.iloc[:,1], percentile_bins)
print("\npercentile_bins:", percentile_bins, "\npercentile_cutoffs", percentile_cutoffs)
```

```python
#label with different pc_change cutoffs
blank = []
for n in [80]:
    bins = [0, n, 100]                         #Note: can adjust here if want to use different labelling
    label = ["0", "1"]                                   #Note: can adjust here if want to use different labelling
    result = run_label_percentile(df, bins, label, 1).reset_index(drop=True)
    blank.append(result['label'])

df_blank = pd.DataFrame(blank).T
final_labels_combinded = pd.concat([df['punk_id'], df_blank], axis=1)
final_labels_combinded.to_csv('mp_final_labels_combined.csv', index=False)
final_labels_combinded
```

```python
#features
df_n = pd.read_csv('ip_df_without_noise.csv')
df_l = final_labels_combinded.copy()
con = df_l['punk_id'].tolist()
df_n = df_n[df_n['id'].isin(con) == True].reset_index(drop=True)

df_n.to_csv('mp_final_features.csv', index=False)
df_n
```
