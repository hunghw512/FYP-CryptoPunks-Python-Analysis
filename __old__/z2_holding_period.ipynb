```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('df_last_two_txn.csv')  
```

```python
df = df.sort_values(['punk_id', 'block_time'], ascending=[True, False])
```

```python
pd.to_datetime(df.block_time)
```

```python
df['previous'] = pd.to_datetime(df.block_time).shift(periods=-1, fill_value=0)
df['holding_period'] = (pd.to_datetime(df.block_time)-df.previous)/np.timedelta64(1, 'M')
df['holding_period'] = df['holding_period']#.astype(int)
df = df.drop_duplicates(subset=['punk_id'], keep='first')
df
```

```python
df = df[df['holding_period'] == 0]
df
```
